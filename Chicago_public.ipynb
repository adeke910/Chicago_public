##### Highest Safety Score

##### School with the highest Safety Score

##### Top 10 schools with the highest "Average Student Attendance"


```python
%sql select NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE from CHICAGO_PUBLIC_SCHOOLS order by Average_Student_Attendance desc nulls last limit 10 
```

#### Five Schools with the lowest Average Student Attendance based on attendance

```python
%sql select NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE from CHICAGO_PUBLIC_SCHOOLS order by Average_Student_Attendance nulls last limit 5
```

##### Eliminating the '%' sign from result set above

```python
%sql select NAME_OF_SCHOOL, replace(AVERAGE_STUDENT_ATTENDANCE, '%','') from CHICAGO_PUBLIC_SCHOOLS order by Average_Student_Attendance nulls last limit 5
```

##### Schools with Average Student Attendance lower than 70%


```python
%sql select NAME_OF_SCHOOL, AVERAGE_STUDENT_ATTENDANCE from CHICAGO_PUBLIC_SCHOOLS where AVERAGE_STUDENT_ATTENDANCE < '70%';
```

##### Total College Enrollment for each Community Area

```python
%sql select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT from CHICAGO_PUBLIC_SCHOOLS group by Community_Area_Name; 
```

##### Five Community Areas with the least total College Enrollment 

```python
%sql select Community_Area_Name, sum(College_Enrollment) AS TOTAL_ENROLLMENT from CHICAGO_PUBLIC_SCHOOLS group by Community_Area_Name order by TOTAL_ENROLLMENT limit 5; 
```
